Line plot

```python
plt.plot(df['Age'].drop.values)
plt.title("Passenger Ages")
plt.xlabel("Passenger Index")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

Bar Chart

```python
survived_counts = df['Survived'].value_counts()

plt.bar(['Not Survived', 'Survived'], survived_counts, color=['red', 'green'])
plt.title("Survival Counts")
plt.ylabel("Number of Passengers")
plt.show()
```

Pie Chart

```python
classes = df['Pclass'].value_counts()
labels = ['Class 1', 'Class 2', 'Class 3']
colors = ['blue', 'red', 'green']

plt.pie(classes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Passenger Class Distribution")
plt.show()
```

Histogram

```python
plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.hist(df['Age'].dropna(), bins=10, alpha=0.5, label='Age', color='skyblue', edgecolor='black')
plt.hist(df['Fare'].dropna(), bins=30, alpha=0.5, label='Fare', color='salmon', edgecolor='black')

plt.title("Age vs Fare Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Box Plot

```python
plt.boxplot(df['Fare'].dropna())
plt.title("Fare Boxplot")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

```python
plt.boxplot([df['Age'].dropna(), df['Fare'].dropna()],
            labels=['Age', 'Fare'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'))

plt.title("Boxplot of Age and Fare")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

Subplots

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].hist(df['Age'].dropna(), bins=15, color='orange')
axs[0].set_title("Age Histogram")

axs[1].boxplot(df['Fare'].dropna())
axs[1].set_title("Fare Boxplot")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].hist(df['Age'].dropna(), color='skyblue')
axs[0, 0].set_title("Age")

axs[0, 1].boxplot(df['Fare'].dropna())
axs[0, 1].set_title("Fare")

axs[1, 0].bar(['Male', 'Female'], df['Sex'].value_counts()[['male', 'female']])
axs[1, 0].set_title("Gender Count")

axs[1, 1].pie(df['Pclass'].value_counts(), labels=['1st', '2nd', '3rd'], autopct='%1.1f%%')
axs[1, 1].set_title("Class Distribution")

plt.tight_layout()
plt.show()
```

Grouped Bar Chart

```python
grouped = df.groupby(['Pclass', 'Survived']).size().unstack()

grouped.plot(kind='bar', stacked=False, color=['red', 'green'])
plt.title("Survival by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(["Not Survived", "Survived"])
plt.show()
```
